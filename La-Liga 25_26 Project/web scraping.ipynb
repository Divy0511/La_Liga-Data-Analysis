```python
import pandas  as pd
```

```python
import requests

def read_wiki_table(url, table_index):
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; MyLaLigaScraper/1.0; +https://yourdomain.example/contact)"
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    tables = pd.read_html(response.text)
    return tables[table_index]

df_24_25 = read_wiki_table("https://en.wikipedia.org/wiki/2024%E2%80%9325_La_Liga", 5)
df_23_24 = read_wiki_table("https://en.wikipedia.org/wiki/2023%E2%80%9324_La_Liga", 5)
df_22_23 = read_wiki_table("https://en.wikipedia.org/wiki/2022%E2%80%9323_La_Liga", 5)
df_21_22 = read_wiki_table("https://en.wikipedia.org/wiki/2021%E2%80%9322_La_Liga", 4)
df_20_21 = read_wiki_table("https://en.wikipedia.org/wiki/2020%E2%80%9321_La_Liga", 4)
df_19_20 = read_wiki_table("https://en.wikipedia.org/wiki/2019%E2%80%9320_La_Liga", 4)
df_18_19 = read_wiki_table("https://en.wikipedia.org/wiki/2018%E2%80%9319_La_Liga", 4)
df_17_18 = read_wiki_table("https://en.wikipedia.org/wiki/2017%E2%80%9318_La_Liga", 4)
df_16_17 = read_wiki_table("https://en.wikipedia.org/wiki/2016%E2%80%9317_La_Liga", 4)
df_15_16 = read_wiki_table("https://en.wikipedia.org/wiki/2015%E2%80%9316_La_Liga", 4)
df_14_15 = read_wiki_table("https://en.wikipedia.org/wiki/2014%E2%80%9315_La_Liga", 4)

dfs = [df_24_25, df_23_24, df_22_23, df_21_22, df_20_21, df_19_20, df_18_19, df_17_18, df_16_17, df_15_16, df_14_15]

for x in dfs:
    x['Pts'] = x['Pts'].astype(str).str.replace(r'[^\d]', '', regex=True).astype(int)



# Strip spaces from all column names
for x in dfs:
    x.columns = x.columns.str.strip()



for x in dfs:
    if 'Teamvte' in x.columns:
        x.rename(columns={'Teamvte': 'Team'}, inplace=True)


# Drop the column if it exists
for i in range(len(dfs)):
    if "Qualification or relegation" in dfs[i].columns:
        dfs[i] = dfs[i].drop(columns=["Qualification or relegation"])

def format_table(df, season_start, season_end):
    print(f"\nSeason {season_start} - {season_end}")
    print("-" * 80)  # Add separator line
    
    # Format column headers
    headers = df.columns.tolist()
    header_format = "{:>3} {:<20} {:>3} {:>3} {:>3} {:>3} {:>4} {:>4} {:>4} {:>4}"
    print(header_format.format(*headers))
    print("-" * 80)  # Add separator line
    
    # Format and print each row
    row_format = "{:>3} {:<20} {:>3} {:>3} {:>3} {:>3} {:>4} {:>4} {:>4} {:>4}"
    for _, row in df.iterrows():
        values = row.tolist()
        # Ensure team name doesn't exceed 20 characters
        values[1] = values[1][:20]
        print(row_format.format(*values))

for x in range(len(dfs)):
    format_table(dfs[x], 24 - x, 25 - x)
```

```python
# Add season information and clean team names
for i, df in enumerate(dfs):
    df['Season'] = f'{24-i}/{25-i}'
    
    # Function to clean special characters
    def clean_text(text):
        replacements = {
            'é': 'e', 'á': 'a', 'í': 'i', 'ó': 'o', 'ñ': 'n',
            'Ã©': 'e', 'Ã¡': 'a', 'Ã­': 'i', 'Ã³': 'o', 'Ã±': 'n'
        }
        for old, new in replacements.items():
            text = text.str.replace(old, new)
        return text

    # Clean special characters in Team column
    if 'Team' in df.columns:
        df['Team'] = clean_text(df['Team'])
    
    if 'Teamvte' in df.columns:
        df['Teamvte'] = clean_text(df['Teamvte'])
    
    # Fix GD (Goal Difference) values
    if df['GD'].dtype == 'object':
        df['GD'] = df['GD'].str.replace('âˆ', '-').str.replace('−', '-').astype(int)
# Concatenate all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

# Save to CSV with UTF-8 encoding
combined_df.to_csv('LaLiga_Seasons_2014_2025.csv', index=False, encoding='utf-8')

print("CSV file has been created successfully with cleaned data!")
```
