```python
full_data['Point_per_match'] = round(full_data['Pts'] / full_data['Pld'], 2)
full_data['Win_percentage'] = round((full_data['W'] / full_data['Pld'] * 100), 2)

print(full_data.to_string(index=False))
```

```python
full_data['Team'] = full_data['Team'].str.replace(r"\(C\)|\(R\)|\[d\] |\[g\]", "", regex=True).str.strip()

total_points_table = full_data.groupby('Team').agg(
    Total_Points=('Pts', 'sum'),
    Average_Points_Per_Match=('Point_per_match', 'mean'),  # Add this line
    Win_Percentage=('Win_percentage', 'mean')
).sort_values(by='Win_Percentage', ascending=False).reset_index()

# Insert a 'Rank' column based on the new index
total_points_table.insert(0, 'rank', total_points_table.index + 1)

# Round the numeric columns for better display
total_points_table['Average_Points_Per_Match'] = total_points_table['Average_Points_Per_Match'].round(2)
total_points_table['Win_Percentage'] = total_points_table['Win_Percentage'].round(2)

print("--- Total performance Points Table 2014-2025 ---")
print(total_points_table.to_string(index=False))

duplicate_teams = total_points_table[total_points_table.duplicated(['Team'], keep=False)]
print("--- Duplicate Teams ---")
print(duplicate_teams)
```

```python
import requests 

def read_wiki_table(url, table_index):
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; MyLaLigaScraper/1.0; +https://yourdomain.example/contact)"
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    tables = pd.read_html(response.text)
    return tables[table_index]

df_25_26 = read_wiki_table("https://en.wikipedia.org/wiki/2025%E2%80%9326_La_Liga", 5)

df_25_26 = df_25_26[["Teamvte"]]
print(df_25_26.to_string(index=False))
```

```python
import unidecode

total_points_table['Team_normalized'] = total_points_table['Team'].astype(str).apply(
    lambda x: unidecode.unidecode(x).strip().lower()
)

df_25_26['Teamvte_normalized'] = df_25_26['Teamvte'].astype(str).apply(
    lambda x: unidecode.unidecode(x).strip().lower()
)


Predicted_table_25_26 = total_points_table[
    total_points_table['Team_normalized'].isin(df_25_26['Teamvte_normalized'])
]

Oviedo_team_row = {'Team': 'Oviedo', 'Predicted_Points': 'Did not play before', 'Predicted_Win_Matches': 'Did not play before'}
Predicted_table_25_26 = pd.concat([Predicted_table_25_26, pd.DataFrame([Oviedo_team_row])], ignore_index=True)


Predicted_table_25_26['Predicted_Win_Matches'] = Predicted_table_25_26['Win_Percentage'] * 38 / 100
Predicted_table_25_26['Predicted_Points'] = ((Predicted_table_25_26['Average_Points_Per_Match'] * 38) + (Predicted_table_25_26['Total_Points'] / 10)) / 2

Predicted_table_25_26 = Predicted_table_25_26[['Team', 'Predicted_Points', 'Predicted_Win_Matches']].sort_values(by='Predicted_Points', ascending=False)

print(Predicted_table_25_26.to_string(index=False))

Predicted_table_25_26.to_csv('Predicted_table_25_26.csv', index=False)
```
